
In this notebook we will take a look at the simple Hello World scenario of TensorFlow Datasets (TFDS). We'll use TFDS to perform the extract, transform, and load processes for the MNIST dataset.

#Importing Tfds

```
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow_datasets as tfds
```

Now tfds works in the form of ETL Pipelines. 


*   for **Extract** method we perform *tfds.load()*
*   for **transform** we use *tfds.shuffle()* here
*   for **load** we use *take* method 



```
dataset = tfds.load(name="mnist", split = "train")   #extract

dataset.shuffle(100)   #transform

for data in dataset.take(1):
  image = data["image"].numpy().squeeze()
  label = data["label"].numpy()

  print("label {}".format(label))
  plt.imshow(image, cmap=plt.cm.binary)
  plt.show()
```
